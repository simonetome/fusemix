```python
import importlib
import fusemix.data_loading as data_loading

from ucimlrepo import fetch_ucirepo

import pandas as pd

importlib.reload(data_loading)
```

```python
dataset_info = pd.read_csv("dataset_ids.csv", sep=" ")

dataset_info['Rows'] = 0
dataset_info['Num_features'] = 0
dataset_info['Num_categoricals'] = 0
dataset_info['Num_classes'] = 0
dataset_info['Missing_features'] = 0
dataset_info['Missing_rows'] = 0
dataset_info['Total_missings'] = 0


dataset_info.head()
```

```python
for id in dataset_info['Id']:
    loaded_dataset = data_loading.load_uciml(id=id, na_drop=False)
    X = loaded_dataset['X']
    dataset_info.loc[dataset_info['Id'] == id, 'Num_classes'] =loaded_dataset['y'].value_counts().shape[0]
    dataset_info.loc[dataset_info['Id'] == id, 'Missing_rows'] = sum(X.isnull().sum(axis = 1) > 0)
    dataset_info.loc[dataset_info['Id'] == id, 'Rows'] = X.shape[0]
    dataset_info.loc[dataset_info['Id'] == id, 'Num_features'] = X.shape[1]
    dataset_info.loc[dataset_info['Id'] == id, 'Num_categoricals'] = len(loaded_dataset['cat_variables'])
    dataset_info.loc[dataset_info['Id'] == id, 'Missing_features']  = (X.isna().sum() > 0).sum()
    dataset_info.loc[dataset_info['Id'] == id, 'Total_missings'] = X.isna().sum().sum()
    
```
