```python
# load simulation results

simulation_path = "../../test_output/sensitivity_results08_01_2026.pkl"

with open(simulation_path, 'rb') as input_file:
    simulation_results = pickle.load(input_file)
```

```python
# Load simulation configuration

with open("../../test_data/simulation_config.yaml", "r") as f:
    cfg = yaml.safe_load(f)

# conf is a tuple (dataset_id,props,mf_proportions,mnar_proportions, seed)

configs = list(product(
    cfg['dataset_ids'],
    cfg['md_param_grid']['props'],
    cfg['md_param_grid']['mf_proportions'],
    cfg['md_param_grid']['mnar_proportions'],
    range(cfg['n_runs'])
    ))

consensuns_thresholds = [0.2,0.3,0.5,0.7]
knn_numbers = [5,10,15,20,25] 

sensitivity_conf = product(knn_numbers, consensuns_thresholds)
```

```python
# simulations_results[conf][(knn, co_thresh)] 

for knn, co_thresh in product(knn_numbers, consensuns_thresholds):
    print(knn)
```

```python
# load configs of a specific dataset 
dataset_ = cfg['dataset_ids'][0]

cfg_dataset = [cfg for cfg in configs if cfg[0] == dataset_]

```

```python
# results of a specific dataset 
result_datasets = {cfg:simulation_results[cfg] for cfg in cfg_dataset}
#
cfg for result_datasets
```

```python
for cfg in result_datasets.values():
    for sns in sensitivity_conf:
        print(cfg[sns]['external_metrics']['ami'])
```
